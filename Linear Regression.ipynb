```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
pageSpeeds = np.random.normal(3.0,0.9,1000)

# Normal Distribution around mean 3 and standard deviation and sd =1   
purchaseAmount = 100 - (pageSpeeds + np.random.normal(0,0.1,1000)) *3
```

```python
plt.scatter(pageSpeeds,purchaseAmount)
```

```python
from scipy import stats
```

```python
stats
```

```python
slope,intercept,rvalue,pvalue ,stderr= stats.linregress(pageSpeeds,purchaseAmount)
```

```python
rvalue ** 2 
```

```python
# Lets use the obtained slop and intercept value to plot a graph containg predicted values and Observed
import matplotlib.pyplot as plt
def predict(x):
    return slope*x + intercept
fitLine = predict(pageSpeeds)
plt.scatter(pageSpeeds,purchaseAmount,c = 'g')
plt.plot(pageSpeeds,fitLine,c = 'r')
plt.show()
```
