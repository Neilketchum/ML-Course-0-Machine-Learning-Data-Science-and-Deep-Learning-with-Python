```python
import numpy as np
np.random.seed(2)
```

```python
import matplotlib.pyplot as plt
pageSpeeds = np.random.normal(3,0.9,1000)
purchaseAmount = np.random.normal(50.0,10.0,1000) / pageSpeeds
```

```python
plt.scatter(pageSpeeds,purchaseAmount)
```

```python
x = np.array(pageSpeeds)
y = np.array(purchaseAmount)
p4 = np.poly1d(np.polyfit(x,y,7)) #Fitting to a degree 4
```

```python
xp = np.linspace(0,7,100) # numpy.linspace(start, stop, num=50, endpoint=True, retstep=False, dtype=None, axis=0)[source]

   # Return evenly spaced numbers over a specified interval.
plt.scatter(x,y,c='y')
plt.plot(xp,p4(xp),c='r')
plt.show()
```

```python
from sklearn.metrics import r2_score
```

```python
r2 = r2_score(y,p4(x))
r2
```
